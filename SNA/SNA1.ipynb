```python
# 2.加载数据文件

import pandas as pd

# 文件路径（请替换为实际路径）
file_path = '/content/zhihunw.xlsx'  # 确保文件已上传到Colab
data = pd.ExcelFile(file_path)

# 查看所有Sheet名称
print("工作表名称: ", data.sheet_names)

# 读取目标Sheet
df = data.parse('Sheet1')
print("数据预览:")
df.head()
```

```python
# 3.准备网络数据

# 提取目标列：被点赞用户和点赞用户列表
liked_users = df.iloc[:, 1]  # B列：被点赞用户
liking_users_raw = df.iloc[:, 5]  # F列：点赞用户列表

# 数据预处理：利用 `_x000d_` 作为分隔符
edges = []
for liked, liking_list in zip(liked_users, liking_users_raw):
    # 检查 liked 和 liking_list 是否为有效数据
    if pd.notna(liked) and pd.notna(liking_list):  # 排除 NaN 值
        # 将 liked 和 liking_list 转换为字符串
        liked = str(liked).strip()
        liking_users = [user.strip() for user in str(liking_list).split('_x000d_')]
        # 构建点赞边列表
        edges.extend([(liked, liker) for liker in liking_users if liker])

# 构建边关系数据框
edges_df = pd.DataFrame(edges, columns=['Liked', 'Liker'])

# 检查结果
print("生成的边关系数据预览:")
print(edges_df.head())

# 保存为 CSV 文件
edges_df.to_csv('edge_list.csv', index=False)
print("边列表已保存为 edge_list.csv")
```

```python
# 4. 网络分析

# 导入必要的库
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 设置中文字体
font_path = "/usr/share/fonts/opentype/noto/NotoSansCJK-Regular.ttc"  # 字体路径
prop = fm.FontProperties(fname=font_path)

# 读取点赞边列表 CSV 文件
edge_list = pd.read_csv('edge_list.csv')  # 确保文件路径正确

# 构建有向图
graph = nx.DiGraph()
for _, row in edge_list.iterrows():
    graph.add_edge(row['Liker'], row['Liked'])

# **1. 节点层面指标计算**
in_degrees = dict(graph.in_degree())  # 被点赞次数
out_degrees = dict(graph.out_degree())  # 点赞次数

# 度中心性、紧密中心性、中介中心性、Pagerank
degree_centrality = nx.degree_centrality(graph)
closeness_centrality = nx.closeness_centrality(graph)
betweenness_centrality = nx.betweenness_centrality(graph)
pagerank = nx.pagerank(graph)

# 保存节点层面指标到文件
node_metrics = pd.DataFrame({
    "Node": list(graph.nodes),
    "InDegree": [in_degrees.get(node, 0) for node in graph.nodes],
    "OutDegree": [out_degrees.get(node, 0) for node in graph.nodes],
    "DegreeCentrality": [degree_centrality.get(node, 0) for node in graph.nodes],
    "ClosenessCentrality": [closeness_centrality.get(node, 0) for node in graph.nodes],
    "BetweennessCentrality": [betweenness_centrality.get(node, 0) for node in graph.nodes],
    "PageRank": [pagerank.get(node, 0) for node in graph.nodes],
})
node_metrics.to_csv("node_metrics.csv", index=False)
print("节点层面指标已保存为 node_metrics.csv")

# **2. 网络层面指标计算**
if nx.is_connected(graph.to_undirected()):
    avg_path_length = nx.average_shortest_path_length(graph)
    diameter = nx.diameter(graph)
    print("平均路径长度:", avg_path_length)
    print("网络直径:", diameter)
else:
    print("网络是非连通的，无法计算平均路径长度和直径。")
    avg_path_length = "N/A"
    diameter = "N/A"

clustering_coefficient = nx.average_clustering(graph.to_undirected())
density = nx.density(graph)

network_metrics = {
    "AveragePathLength": avg_path_length,
    "Diameter": diameter,
    "ClusteringCoefficient": clustering_coefficient,
    "Density": density,
}
pd.DataFrame([network_metrics]).to_csv("network_metrics.csv", index=False)
print("网络整体指标已保存为 network_metrics.csv")

# **3. 网络可视化**
filtered_nodes = [node for node, out_degree in graph.out_degree() if out_degree > 10]
subgraph = graph.subgraph(filtered_nodes)

plt.figure(figsize=(12, 8))
pos = nx.spring_layout(subgraph, k=0.5, iterations=50)

# 绘制节点和边
nx.draw_networkx_edges(subgraph, pos, edge_color="gray", alpha=0.7)
nx.draw_networkx_nodes(subgraph, pos, node_color="skyblue", node_size=50)

# 手动绘制节点标签，使用 Matplotlib 添加
for node, (x, y) in pos.items():
    plt.text(
        x, y, node, fontsize=8, fontproperties=prop, ha="center", va="center"
    )

# 设置标题
plt.title("点赞网络图", fontproperties=prop, fontsize=16)
plt.axis("off")
plt.show()

# 保存简化后的网络
nx.write_gexf(subgraph, "simplified_social_network.gexf")
print("简化的网络已保存为 simplified_social_network.gexf")
```
